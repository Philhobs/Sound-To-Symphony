# Autoenc Model


```python
#auto encoder model, lifted straight from the tensorflow site.

class Autoencoder(Model):
    def __init__(self, latent_dim, shape):
        super(Autoencoder, self).__init__()
        self.latent_dim = latent_dim
        self.shape = shape
        self.encoder = tf.keras.Sequential([
            
          
          layers.Conv1D(6, 12, activation='relu', input_shape=(700,33)),
          #layers.Conv1D(3, 6, activation='relu'),
          
          #layers.Normalization(axis=None),
          #layers.Dense(1000, activation='linear'),
          #layers.Dense(25, activation='relu'), #relu?
          layers.Flatten(),
          layers.Dense(latent_dim, activation='relu'), #relu?
        ])
        
        self.decoder = tf.keras.Sequential([
          #layers.Normalization(axis=None, invert=True),
          #layers.Conv1DTranspose(16, 6, activation='relu'),
          #layers.Conv1DTranspose(32, 3, activation='relu'),
          layers.Dense(tf.math.reduce_prod(shape), trainable=True, activation = 'linear'),
          layers.Reshape(shape),
        ])
    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

## Next train model, display output

```python
# training data from 3.2
X_train_stack = X_train_mels_db
```

```python
X_train_stack.shape
```

```python
shape = X_train_stack[0].shape

# the 'bottle neck' of our neural network.
latent_dim = 300

autoencoder = Autoencoder(latent_dim, shape)

autoencoder.compile(optimizer='adam', loss=losses.MeanAbsoluteError())
```

```python
autoencoder.fit(X_train_stack, 
                X_train_stack, 
                epochs=100, 
                batch_size = 64,
                shuffle=True, 
                verbose = 1)
```

```python
#autoencoder.encoder.summary() ; autoencoder.decoder.summary()
```

```python
test_noise = np.random.normal(loc=2, scale=1, size=(1,300)) #.reshape(1,50)
#test_noise = sample_x_raw.reshape(1,300)
```

```python
decoded_mel = autoencoder.decoder(test_noise).numpy().reshape(700,33)
```

```python
plt.imshow(decoded_mel.reshape(100,33*7))
```

```python
base_sample = librosa.feature.inverse.mel_to_audio(decoded_mel,  n_fft=2500, sr=SAMPLE_RATE)

ipd.Audio(base_sample, rate=SAMPLE_RATE)
```
